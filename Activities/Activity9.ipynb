```python
m = 1

g = 9.8

b = 1

theta0 = 2

y0 = 0

v0 = np.array([theta0, y0])

dt = 0.01

t = np.arange(0, 5 + dt, dt)

tspan = (0, 5)

f = lambda t, v: np.array([v[1], -m*g*np.sin(v[0]) - b*v[1]])

sol = scipy.integrate.solve_ivp(f, tspan, v0, t_eval=t)

theta = sol.y[0, :]

y = sol.y[1, :]

# Plot

plt.plot(theta, y, 'k')
plt.title("Phase Plane - version 1")
plt.xlabel("theta")
plt.ylabel("y")

# Add lines

x1 = np.array([-1.5, 2])
y1 = np.array([0, 0])
x = np.array([0, 0])
y = np.array([-5, 3])
plt.plot(x, y, 'k--', x1, y1, 'k--')

# Add lines

f1 = lambda theta, y: y
f2 = lambda theta, y: -m*g*np.sin(theta) - b*y
theta = np.linspace(-1.5, 2, 30)
y = np.linspace(-5, 3, 30)
[Theta, Y] = np.meshgrid(theta, y)
plt.quiver(Theta, Y, f1(Theta, Y), f2(Theta, Y))
plt.show()
```
