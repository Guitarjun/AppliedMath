```python
# Problem 1

x = np.arange(-2, 2.25, 0.25)
y = 2*x + 3
plt.figure()
plt.plot(x, y, 'k')
plt.plot(x, y, 'bo')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of a line')
plt.show()

"""
a) plt.plot(x, y, 'bo') creates the blue circle intervals, and plt.plot(x, y, 'k') creates the line.

b) The blue circles represent point intervals on the line y = 2x+3 at every 0.25 interval.
    They are a set of points from -2 to 2.25 with 0.25 intervals.
"""
```
