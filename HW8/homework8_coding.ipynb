```python
# Problem 1

x_true = lambda t : .5 * (np.cos(t) + np.sin(t) + np.exp(-t))

# a) Forwards Euler

x0 = 1
dt = 0.1
t = np.arange(0, 10 + dt, dt)

n = t.size
x = np.zeros(n)
x[0] = x0
for k in range(n-1):
    x[k + 1] = x[k] + dt * (np.cos(t[k]) - x[k])
    
A1 = x.reshape(1, n)

# b) Forwards Euler Error

A2 = np.zeros((1, n))
for k in range(n):
    A2[0, k] = np.abs(x[k] - x_true(t[k]))

# c) Backwards Euler

for k in range(n-1):
    x[k + 1] = (x[k] + dt * (np.cos(t[k+1]))) / (1+dt)
    
A3 = x.reshape(1, n)

# d) Backwards Euler Error

A4 = np.zeros((1, n))
for k in range(n):
    A4[0, k] = np.abs(x[k] - x_true(t[k]))
    
# e) solve_ivp

f = lambda t, x : np.cos(t) - x
tspan = (0, 10)
x0 = np.array([1])
sol = scipy.integrate.solve_ivp(f, tspan, x0, t_eval=t)
x = sol.y[0, :]

A5 = x.reshape(1, n)

# f) solve_ivp error

A6 = np.zeros((1, n))
for k in range(n):
    A6[0, k] = np.abs(x[k] - x_true(t[k]))
```
